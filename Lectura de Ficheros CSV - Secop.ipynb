```python
import csv
```

```python
with open('data/secopi_2015.csv', encoding='latin1') as fichero_csv:
    
    lector = csv.reader(fichero_csv)
    next(lector, None)
    secop = []
    for indice, linea in enumerate(lector):
        proceso = {}
        proceso["numero_constancia"] = linea[0]
        proceso["tipo_proceso"] = linea[2]
        proceso["nombre_entidad"] = linea[3]
        proceso["estado_proceso"] = linea[6]
        if linea[2] == "Licitación Pública":
            unspsc = {}
            unspsc["codigo"] = linea[7]
            unspsc["nombre"] = linea[8]
            proceso["unspsc"] = unspsc
            secop.append(proceso)
            
        if indice == 10000:
            break
       
    print(secop)
```
