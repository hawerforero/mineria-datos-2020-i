```python
#Redes Neuronales

data_training = [{"input":[1,0,1], "output":1 },
                 {"input":[0,1,1], "output":0 },
                 {"input":[1,1,1], "output":1 },
                 {"input":[0,0,1], "output":0 },
                 {"input":[0,1,0], "output":0 }]

print(data_training)
```

```python
#Sumatoria
def _sum(_input, weight):
    summation = 0
    for index, i in enumerate(_input):
        summation += i * weight[index]
    return summation

print(_sum(data_training[0]['input'], weight))
```

```python
#Sigmoide
import math
def _sigmoide(summation):
    
    resultado = 1 / (1 + math.exp(-summation))
    return resultado

print(_sigmoide(0.4060574446522389))
```

```python
#Gradiente
def _gradient(y):
    resultado = y * (1-y)
    return resultado
print(_gradient(0.6001421523462186))
```

```python
def _error(output, y):
    resultado = output - y
    return resultado
print(_error(1, 0.6001421523462186))
```

```python
def _weightAdjustment(xi, error, gradient):
    return xi * error * gradient

```

```python
def _outPutRNA(_input):
    summation = _sum(_input, weight)
    y = _sigmoide(summation)
    return y
    
```

```python
#Pesos iniciales (Aleatorio -1, 1)
import random 

weight = [random.uniform(-1, 1), random.uniform(-1, 1), random.uniform(-1, 1)]
                         
        
print(weight)
```

```python
#training
epocas = 100

for i in range(epocas):
    print("**EPOCA** ",i)
    
    for data in data_training:
        summation = _sum(data['input'], weight)
        y = _sigmoide(summation)
        print("Y-> ", y, " Output-> ",data['output'])

        gradient = _gradient(y)
        error = _error(data['output'], y)

        #Ajustar los pesos 
        for index, w in enumerate(weight):
            weight[index] += _weightAdjustment(data['input'][index], error, gradient)
```

```python
#Output

print(_outPutRNA([1,1,0]))
```
